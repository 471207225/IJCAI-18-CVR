```python
import pandas as pd
import gc
from sklearn import preprocessing
import time
import math
import pickle
import numpy as np
```

```python
total =  pickle.load(open('../../data/raw/total_high_level.pkl','rb+'))
total = total.reset_index(drop=True)
```

```python
data = total.loc[(total["context_timestamp"] >= "20180907"),["user_id","user_id_shop_id","user_id_item_id","context_timestamp","is_trade"]]
```

```python
data["apperance"] = 1
```

```python
feature_name = "user_id"
data.sort_values(by=[feature_name, 'context_timestamp'], inplace=True)
data['apperance_nums_{}'.format(feature_name)]=data.groupby([feature_name])["apperance"].cumsum()  
df = data.groupby([feature_name])["apperance"].sum()  
result = pd.DataFrame({feature_name:df.index, 'apperance_nums_sum_{}'.format(feature_name):df.values})
data = data.reset_index().merge(result,on=feature_name, how="left").set_index("index")
data["apperance_nums_{}_ratio".format(feature_name)] = data['apperance_nums_{}'.format(feature_name)] / data['apperance_nums_sum_{}'.format(feature_name)]
del data['apperance_nums_{}'.format(feature_name)]
del data['apperance_nums_sum_{}'.format(feature_name)]
```

```python
feature_name = "user_id_item_id"
data.sort_values(by=[feature_name, 'context_timestamp'], inplace=True)
data['apperance_nums_{}'.format(feature_name)]=data.groupby([feature_name])["apperance"].cumsum()  
df = data.groupby([feature_name])["apperance"].sum()  
result = pd.DataFrame({feature_name:df.index, 'apperance_nums_sum_{}'.format(feature_name):df.values})
data = data.reset_index().merge(result,on=feature_name, how="left").set_index("index")
data["apperance_nums_{}_ratio".format(feature_name)] = data['apperance_nums_{}'.format(feature_name)] / data['apperance_nums_sum_{}'.format(feature_name)]
del data['apperance_nums_{}'.format(feature_name)]
del data['apperance_nums_sum_{}'.format(feature_name)]
```

```python
feature_name = "user_id_shop_id"
data.sort_values(by=[feature_name, 'context_timestamp'], inplace=True)
data['apperance_nums_{}'.format(feature_name)]=data.groupby([feature_name])["apperance"].cumsum()  
df = data.groupby([feature_name])["apperance"].sum()  
result = pd.DataFrame({feature_name:df.index, 'apperance_nums_sum_{}'.format(feature_name):df.values})
data = data.reset_index().merge(result,on=feature_name, how="left").set_index("index")
data["apperance_nums_{}_ratio".format(feature_name)] = data['apperance_nums_{}'.format(feature_name)] / data['apperance_nums_sum_{}'.format(feature_name)]
del data['apperance_nums_{}'.format(feature_name)]
del data['apperance_nums_sum_{}'.format(feature_name)]
```

```python
data = data.sort_index()
```

```python
del data["user_id"]
del data["user_id_shop_id"]
del data["user_id_item_id"]
del data["context_timestamp"]
del data["apperance"]
```

```python
data.to_pickle('../../data/processed/magric_feature.pkl')
```
